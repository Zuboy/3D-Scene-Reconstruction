<h2 style="color:#007b90">Provided Utility Functions</h2>

These functions are provided to reduce the complexity of some steps you might encounter. They were involved in the creation of the given samples. However, you do not have to use them and can use other means of achieving the same results.

```python
def sample_image(numpy_image, numpy_sample_grid):
    '''
    This function samples a target image from a source image (numpy_image) based on specified pixel coordinates (numpy_sample_grid).
    Inputs:
        numpy_image: of shape=[H, W, C]. H is the height, W is the width, and C is the color channel of the source image from which color values will be sampled.
        numpy_sample_grid: of shape=[H, W, UV]. H is the height and W is the width of the target image that will be sampled. UV are the pixel locations in the source image from which to sample color values.
    Outputs:
        sampled_image: of shape=[H, W, C]. H is the height, W is the width, and C is the color channel of the target image that was sampled.
    
    The following is a simple toy example to see the behavior of the sample_image function

    example_source = (np.random.rand(110, 220, 3)*255).astype(int) #[110, 220, 3]
    example_grid = np.ones([100, 200, 2]) #[100, 200, 2]
    example_grid[:, :, 1] = 2
    example_target = sample_image(example_source, example_grid) #[100, 200, 3]

    example_target will be of shape [100, 200, 3]
    -> 100, because example_grid has a height of 100
    -> 200, because example_grid has a width of 200
    -> 3, because example_source has a color channel of 3
    print(example_source.shape)

    example_target will contain the value of example_source[2, 1] at every pixel
    -> 2, because example_grid[:, :, 1] has a value of 2 for every pixel
    -> 1, because example_grid[:, :, 0] has a value of 1 for every pixel
    print(example_source[2, 1, 0], "->", np.unique(example_target[:, :, 0]))
    print(example_source[2, 1, 1], "->", np.unique(example_target[:, :, 1]))
    print(example_source[2, 1, 2], "->", np.unique(example_target[:, :, 2]))
    
    '''
    height, width, _ = numpy_image.shape#[H, W, 3]

    #turn numpy array to torch tensor
    torch_sample_grid = torch.from_numpy(numpy_sample_grid)#[H, W, 2]
    #normalize from range (0, width-1) to (0, 1)
    torch_sample_grid[:, :, 0] = torch_sample_grid[:, :, 0] / (width-1)
    #normalize from range (0, height-1) to (0, 1)
    torch_sample_grid[:, :, 1] = torch_sample_grid[:, :, 1] / (height-1)
    #normalize from range (0, 1) to (-1, 1)
    torch_sample_grid = torch_sample_grid*2 -1

    #transform to necessary shapes
    torch_sample_grid = torch_sample_grid.unsqueeze(0)#[1, H, W, 2]
    torch_image = torch.from_numpy(numpy_image).double().permute(2, 0, 1).unsqueeze(0)#[1, 3, H, W]
    #sample image according to sample grid locations from source image
    sampled_image = torch.nn.functional.grid_sample(torch_image, torch_sample_grid, mode='bilinear', padding_mode='zeros', align_corners=True)
    #transform back to numpy image
    sampled_image = sampled_image.squeeze().permute(1, 2, 0).numpy().astype(np.uint8)#[H, W, 3]
    return sampled_image

def ply_creator(input_3d, rgb_data=None, filename='dummy'):
    ''' Creates a colored point cloud that you can visualise using e.g. Meshlab.
    Inputs:
        input_3d: of shape=[N, 3], each row is the 3D coordinate of each point
        rgb_data(optional): of shape=[N, 3], each row is the rgb color value of each point
        filename: file name for the .ply file to be created 
    '''
    assert (input_3d.ndim==2),"Pass 3d points as NumPointsX3 array "
    pre_text1 = """ply\nformat ascii 1.0"""
    pre_text2 = "element vertex "
    pre_text3 = """property float x\nproperty float y\nproperty float z\nproperty uchar red\nproperty uchar green\nproperty uchar blue\nend_header"""
    pre_text22 = pre_text2 + str(input_3d.shape[0])
    pre_text11 = pre_text1
    pre_text33 = pre_text3
    filename = filename + '.ply'
    fid = open(filename, 'w')
    fid.write(pre_text11)
    fid.write('\n')
    fid.write(pre_text22)
    fid.write('\n')
    fid.write(pre_text33)
    fid.write('\n')
    for i in range(input_3d.shape[0]):
        for c in range(3):
            fid.write(str(input_3d[i,c]) + ' ')
        if not rgb_data is None:
            for c in range(3):
                fid.write(str(rgb_data[i,c]) + ' ')
        if i!=input_3d.shape[0]:
            fid.write('\n')
    fid.close()
    return filename 
```

<h2 style="color:#007b90">Step 1: Undistortion of the RGB Images</h2>

As described in the Given Data section, the RGB images contain lens distortion, while the depth images are already undistorted. To ensure accurate 3D reconstruction, the two modalities must be aligned.

Undistort all RGB images, then visualize one example image before and after undistortion.

```python
# Include the necessary code here
def undistort_images(img_path, camera_calibration):
    """
    Manually undistort all RGB images using camera calibration parameters.
    
    Returns:
        undistorted_images: list of undistorted images
        original_images: list of original distorted images (for comparison)
    """
    # Extract calibration parameters
    fx = camera_calibration['focal_length_px']
    fy = fx * camera_calibration['pixel_ratio']
    cx = camera_calibration['principal_point'][1]  # column (x)
    cy = camera_calibration['principal_point'][0]  # row (y)
    
    # Distortion coefficients [k1, k2, p1, p2, k3]
    k1, k2, p1, p2, k3 = camera_calibration['distortion_param']
    
    h = camera_calibration['image_height']
    w = camera_calibration['image_width']
    
    # Create coordinate grids for the undistorted image
    u, v = np.meshgrid(np.arange(w), np.arange(h))
    
    # Convert pixel coordinates to normalized coordinates
    x_norm = (u - cx) / fx
    y_norm = (v - cy) / fy
    
    # Calculate radius squared
    r2 = x_norm**2 + y_norm**2
    r4 = r2**2
    r6 = r2**3
    
    # Apply radial distortion
    radial_distortion = 1 + k1 * r2 + k2 * r4 + k3 * r6
    
    # Apply tangential distortion
    x_distorted = x_norm * radial_distortion + 2 * p1 * x_norm * y_norm + p2 * (r2 + 2 * x_norm**2)
    y_distorted = y_norm * radial_distortion + p1 * (r2 + 2 * y_norm**2) + 2 * p2 * x_norm * y_norm
    
    # Convert back to pixel coordinates
    u_distorted = x_distorted * fx + cx
    v_distorted = y_distorted * fy + cy
    
    # Create the sampling grid [H, W, 2] where [:,:,0] is u (column) and [:,:,1] is v (row)
    sample_grid = np.stack([u_distorted, v_distorted], axis=-1).astype(np.float64)
    
    # Load and undistort all images
    undistorted_images = []
    original_images = []
    
    for i in range(9):
        img_file = os.path.join(img_path, f'{i:02d}.png')
        img = cv.imread(img_file)
        original_images.append(img)
        
        # Sample from the distorted image at the computed locations
        undistorted = sample_image(img, sample_grid)
        undistorted_images.append(undistorted)
    
    return undistorted_images, original_images

# Run undistortion
undistorted_images, original_images = undistort_images(img_path, camera_calibration)

# Visualize one example (e.g., image 0)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(cv.cvtColor(original_images[0], cv.COLOR_BGR2RGB))
axes[0].set_title('Original (Distorted)')
axes[0].axis('off')

axes[1].imshow(cv.cvtColor(undistorted_images[0], cv.COLOR_BGR2RGB))
axes[1].set_title('Undistorted')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print(f"Undistorted {len(undistorted_images)} images successfully.")
```

<h2 style="color:#007b90">Step 2: Common Camera Reference Frame</h2>

The provided camera poses are expressed relative to consecutive frames (e.g. the first pose is a transformation from the first view to the second view). However, the final point cloud needs to be definied with respect to a single, common world coordinate system.

Compute camera poses such that they are expressed in terms of a shared world reference frame, and then visualize all camera positions and orientations in a single 3D plot.

```python
def compute_world_to_camera_poses(camera_movement):
    """
    Compute world-to-camera transformation matrices.
    
    camera_movement[i] transforms FROM camera i TO camera i+1.
    We need the world-to-camera matrices where camera 0 is at world origin.
    
    Returns:
        T_world_to_cam: [9, 4, 4] - transforms points from world to each camera
    """
    num_cameras = camera_movement.shape[0] + 1
    T_world_to_cam = np.zeros((num_cameras, 4, 4))
    
    # Camera 0 is at world origin
    T_world_to_cam[0] = np.eye(4)
    
    # Build transformations
    # T_world_to_cam[i] = camera_movement[i-1] @ T_world_to_cam[i-1]
    for i in range(1, num_cameras):
        T_world_to_cam[i] = camera_movement[i-1] @ T_world_to_cam[i-1]
    
    return T_world_to_cam


def get_camera_centers(T_world_to_cam):
    """
    Extract camera centers from world-to-camera transforms.
    Camera center C satisfies: T_w2c @ [C, 1]^T = [0, 0, 0, 1]^T
    So: C = -R^T @ t where T_w2c = [R|t]
    """
    num_cameras = len(T_world_to_cam)
    camera_centers = np.zeros((num_cameras, 3))
    
    for i in range(num_cameras):
        R = T_world_to_cam[i][:3, :3]
        t = T_world_to_cam[i][:3, 3]
        camera_centers[i] = -R.T @ t
    
    return camera_centers


def visualize_camera_setup(camera_centers, T_world_to_cam):
    """
    Visualize camera positions and orientations.
    """
    from mpl_toolkits.mplot3d import Axes3D
    
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot camera centers
    ax.scatter(camera_centers[:, 0], camera_centers[:, 1], camera_centers[:, 2],
              c='red', marker='o', s=150, edgecolors='black', linewidths=2,
              label='Camera Centers', zorder=5)
    
    # Plot camera path
    ax.plot(camera_centers[:, 0], camera_centers[:, 1], camera_centers[:, 2],
           'r--', linewidth=2, alpha=0.6)
    
    # Plot viewing directions
    for i in range(len(camera_centers)):
        center = camera_centers[i]
        R = T_world_to_cam[i][:3, :3]
        
        # Viewing direction (camera looks along +Z in camera coords)
        # In world: view_dir = -R^T @ [0, 0, 1]
        view_dir = -R.T @ np.array([0, 0, 0.3])
        
        ax.quiver(center[0], center[1], center[2],
                 view_dir[0], view_dir[1], view_dir[2],
                 color='blue', arrow_length_ratio=0.3, linewidth=2)
        
        ax.text(center[0], center[1], center[2], f'  {i}', fontsize=10, weight='bold')
    
    ax.set_xlabel('X (m)')
    ax.set_ylabel('Y (m)')
    ax.set_zlabel('Z (m)')
    ax.set_title('Camera Positions in World Frame')
    ax.legend()
    
    plt.tight_layout()
    plt.show()


print("\n" + "="*60)
print("STEP 2: Computing Camera Poses")
print("="*60)

# Try INVERSE transformations
T_world_to_cam = compute_world_to_camera_poses(np.linalg.inv(camera_movement))
camera_centers = get_camera_centers(T_world_to_cam)

print(f"\nCamera centers in world coordinates:")
for i, center in enumerate(camera_centers):
    print(f"  Camera {i}: [{center[0]:7.3f}, {center[1]:7.3f}, {center[2]:7.3f}]")

print(f"\nCamera center spread:")
cam_span = camera_centers.max(axis=0) - camera_centers.min(axis=0)
print(f"  X span: {cam_span[0]:.3f} m")
print(f"  Y span: {cam_span[1]:.3f} m")
print(f"  Z span: {cam_span[2]:.3f} m")

visualize_camera_setup(camera_centers, T_world_to_cam)
```

<h2 style="color:#007b90">Step 3: Feature Matching and Depth Map Scaling</h2>

Each depth map is scaled differently according to its own maximum value. To obtain a consistent dense point cloud, we must correct these scalings so that 3D points from different views align properly.

Perform feature matching across the images, and use the corresponding matches to determine the correct scale factors for the depth maps.
Additionally, visualize one selected feature across all views.

```python

# ============================================================================
# STEP 3: Depth Scaling Using Triangulation
# ============================================================================

def triangulate_point_from_two_views(pt1, pt2, T1, T2, K):
    """
    Triangulate a 3D point from two 2D observations.
    
    Args:
        pt1, pt2: 2D points [row, col] in image 1 and 2
        T1, T2: world-to-camera transforms [4x4]
        K: camera intrinsic matrix [3x3]
    
    Returns:
        point_3d: 3D point in world coordinates [3]
    """
    # Convert to normalized coordinates
    fx, fy = K[0, 0], K[1, 1]
    cx, cy = K[0, 2], K[1, 2]
    
    x1 = (pt1[1] - cx) / fx
    y1 = (pt1[0] - cy) / fy
    
    x2 = (pt2[1] - cx) / fx
    y2 = (pt2[0] - cy) / fy
    
    # Projection matrices: P = K @ T_w2c = K @ [R|t]
    P1 = K @ T1[:3, :]
    P2 = K @ T2[:3, :]
    
    # Build linear system A @ X = 0
    A = np.array([
        x1 * P1[2, :] - P1[0, :],
        y1 * P1[2, :] - P1[1, :],
        x2 * P2[2, :] - P2[0, :],
        y2 * P2[2, :] - P2[1, :]
    ])
    
    # Solve using SVD
    _, _, Vt = np.linalg.svd(A)
    X = Vt[-1, :]
    X = X / X[3]  # Homogeneous to 3D
    
    return X[:3]


def compute_depth_scales_triangulation(given_features, depth_path, T_world_to_cam, 
                                      camera_calibration):
    """
    Compute depth scale for each image using triangulation.
    
    For each feature:
    1. Triangulate its 3D position using two views (without depth)
    2. Compare with depth-based reconstruction to find scale
    """
    fx = camera_calibration['focal_length_px']
    fy = fx * camera_calibration['pixel_ratio']
    cx = camera_calibration['principal_point'][1]
    cy = camera_calibration['principal_point'][0]
    
    K = np.array([[fx, 0, cx],
                  [0, fy, cy],
                  [0, 0, 1]])
    
    # Load depth images
    depth_images = []
    for i in range(9):
        depth_file = os.path.join(depth_path, f'{i:02d}.png')
        depth = cv.imread(depth_file, cv.IMREAD_GRAYSCALE).astype(np.float64) / 255.0
        depth_images.append(depth)
    
    # Initialize scales
    depth_scales = np.ones(9)
    scale_estimates = [[] for _ in range(9)]
    
    # For each feature, triangulate and compute scales
    num_features = given_features.shape[1]
    
    for feat_idx in range(num_features):
        # Find cameras where this feature is visible
        visible_cams = []
        for cam_idx in range(9):
            if given_features[cam_idx, feat_idx, 0] != -1:
                visible_cams.append(cam_idx)
        
        if len(visible_cams) < 2:
            continue
        
        # Use first two views to triangulate
        cam_i = visible_cams[0]
        cam_j = visible_cams[1]
        
        pt_i = given_features[cam_i, feat_idx]
        pt_j = given_features[cam_j, feat_idx]
        
        # Triangulate 3D point
        try:
            point_3d_world = triangulate_point_from_two_views(
                pt_i, pt_j, T_world_to_cam[cam_i], T_world_to_cam[cam_j], K
            )
        except:
            continue
        
        # For each camera that sees this feature, compute what scale would match
        for cam_idx in visible_cams:
            pt = given_features[cam_idx, feat_idx]
            row, col = int(pt[0]), int(pt[1])
            
            depth_val = depth_images[cam_idx][row, col]
            if depth_val < 0.01:
                continue
            
            # Transform 3D point to this camera
            T = T_world_to_cam[cam_idx]
            point_cam = T[:3, :3] @ point_3d_world + T[:3, 3]
            
            # True depth is point_cam[2]
            # Measured depth (normalized) is depth_val
            # Scale such that: depth_val * scale = point_cam[2]
            scale = point_cam[2] / depth_val
            
            if 0.1 < scale < 20:  # Sanity check
                scale_estimates[cam_idx].append(scale)
    
    # Compute median scale for each camera
    for cam_idx in range(9):
        if len(scale_estimates[cam_idx]) > 0:
            depth_scales[cam_idx] = np.median(scale_estimates[cam_idx])
    
    print(f"\nDepth scales computed (per camera):")
    for i, scale in enumerate(depth_scales):
        print(f"  Camera {i}: {scale:.6f} ({len(scale_estimates[i])} estimates)")
    
    return depth_scales


print("\n" + "="*60)
print("STEP 3: Computing Depth Scales")
print("="*60)

depth_scales = compute_depth_scales_triangulation(
    given_features, depth_path, T_world_to_cam, camera_calibration
)


# ============================================================================
# STEP 4: Depth to 3D World Points
# ============================================================================

def depth_to_world_points_final(undistorted_images, depth_path, T_world_to_cam,
                                depth_scales, camera_calibration):
    """
    Convert depth maps to 3D world points.
    
    For each pixel:
    1. Unproject to 3D in camera coordinates using scaled depth
    2. Transform to world: P_world = T_cam_to_world @ P_cam
       where T_cam_to_world = inv(T_world_to_cam)
    """
    fx = camera_calibration['focal_length_px']
    fy = fx * camera_calibration['pixel_ratio']
    cx = camera_calibration['principal_point'][1]
    cy = camera_calibration['principal_point'][0]
    
    all_points = []
    all_colors = []
    
    for cam_idx in range(9):
        print(f"Processing camera {cam_idx}/8...")
        
        # Load depth
        depth_file = os.path.join(depth_path, f'{cam_idx:02d}.png')
        depth_img = cv.imread(depth_file, cv.IMREAD_GRAYSCALE).astype(np.float64) / 255.0
        depth_img = depth_img * depth_scales[cam_idx]
        
        rgb_img = undistorted_images[cam_idx]
        h, w = depth_img.shape
        
        # Pixel coordinates
        u, v = np.meshgrid(np.arange(w), np.arange(h))
        
        # Valid depths
        valid = depth_img > 0.01
        
        # Unproject to camera coordinates
        Z = depth_img[valid]
        X = (u[valid] - cx) * Z / fx
        Y = (v[valid] - cy) * Z / fy
        
        # Homogeneous coordinates
        points_cam_hom = np.stack([X, Y, Z, np.ones_like(X)], axis=0)  # [4, N]
        
        # Transform to world
        T_cam_to_world = np.linalg.inv(T_world_to_cam[cam_idx])
        points_world_hom = T_cam_to_world @ points_cam_hom  # [4, N]
        points_world = points_world_hom[:3, :].T  # [N, 3]
        
        # Get colors
        colors = rgb_img[valid][:, [2, 1, 0]]  # BGR to RGB
        
        all_points.append(points_world)
        all_colors.append(colors)
    
    all_points = np.vstack(all_points)
    all_colors = np.vstack(all_colors)
    
    print(f"\nTotal points: {all_points.shape[0]:,}")
    print(f"\nPoint cloud bounds:")
    print(f"  X: [{all_points[:, 0].min():.2f}, {all_points[:, 0].max():.2f}] m")
    print(f"  Y: [{all_points[:, 1].min():.2f}, {all_points[:, 1].max():.2f}] m")
    print(f"  Z: [{all_points[:, 2].min():.2f}, {all_points[:, 2].max():.2f}] m")
    
    return all_points, all_colors


print("\n" + "="*60)
print("STEP 4: Depth to 3D World Points")
print("="*60)

all_points_3d, all_colors = depth_to_world_points_final(
    undistorted_images, depth_path, T_world_to_cam,
    depth_scales, camera_calibration
)
```

<h2 style="color:#007b90">Step 4: Depth Map to 3D Points</h2>

Using the rescaled depth maps together with the camera poses from Step 2, compute the 3D coordinates for all pixel in each RGB image.

```python
# ============================================================================
# STEP 4: Depth to 3D World Points
# ============================================================================

def depth_to_world_points_final(undistorted_images, depth_path, T_world_to_cam,
                                depth_scales, camera_calibration):
    """
    Convert depth maps to 3D world points.
    
    For each pixel:
    1. Unproject to 3D in camera coordinates using scaled depth
    2. Transform to world: P_world = T_cam_to_world @ P_cam
       where T_cam_to_world = inv(T_world_to_cam)
    """
    fx = camera_calibration['focal_length_px']
    fy = fx * camera_calibration['pixel_ratio']
    cx = camera_calibration['principal_point'][1]
    cy = camera_calibration['principal_point'][0]
    
    all_points = []
    all_colors = []
    
    for cam_idx in range(9):
        print(f"Processing camera {cam_idx}/8...")
        
        # Load depth
        depth_file = os.path.join(depth_path, f'{cam_idx:02d}.png')
        depth_img = cv.imread(depth_file, cv.IMREAD_GRAYSCALE).astype(np.float64) / 255.0
        depth_img = depth_img * depth_scales[cam_idx]
        
        rgb_img = undistorted_images[cam_idx]
        h, w = depth_img.shape
        
        # Pixel coordinates
        u, v = np.meshgrid(np.arange(w), np.arange(h))
        
        # Valid depths
        valid = depth_img > 0.01
        
        # Unproject to camera coordinates
        Z = depth_img[valid]
        X = (u[valid] - cx) * Z / fx
        Y = (v[valid] - cy) * Z / fy
        
        # Homogeneous coordinates
        points_cam_hom = np.stack([X, Y, Z, np.ones_like(X)], axis=0)  # [4, N]
        
        # Transform to world
        T_cam_to_world = np.linalg.inv(T_world_to_cam[cam_idx])
        points_world_hom = T_cam_to_world @ points_cam_hom  # [4, N]
        points_world = points_world_hom[:3, :].T  # [N, 3]
        
        # Get colors
        colors = rgb_img[valid][:, [2, 1, 0]]  # BGR to RGB
        
        all_points.append(points_world)
        all_colors.append(colors)
    
    all_points = np.vstack(all_points)
    all_colors = np.vstack(all_colors)
    
    print(f"\nTotal points: {all_points.shape[0]:,}")
    print(f"\nPoint cloud bounds:")
    print(f"  X: [{all_points[:, 0].min():.2f}, {all_points[:, 0].max():.2f}] m")
    print(f"  Y: [{all_points[:, 1].min():.2f}, {all_points[:, 1].max():.2f}] m")
    print(f"  Z: [{all_points[:, 2].min():.2f}, {all_points[:, 2].max():.2f}] m")
    
    return all_points, all_colors


print("\n" + "="*60)
print("STEP 4: Depth to 3D World Points")
print("="*60)

all_points_3d, all_colors = depth_to_world_points_final(
    undistorted_images, depth_path, T_world_to_cam,
    depth_scales, camera_calibration
)
```

<h2 style="color:#007b90">Step 5: 3D Points to Point Cloud</h2>

Finally, use the provided `ply_creator` function to generate a merged colored point cloud from the computed 3D points from all images. Visualize the resulting point cloud in a viewer such as MeshLab.

```python
# ============================================================================
# STEP 5: 3D Points to Point Cloud (PLY Export)
# ============================================================================

def export_point_cloud_to_ply(all_points_3d, all_colors, output_path, filename='dense_reconstruction'):
    """
    Export the dense 3D point cloud to PLY format.
    
    Args:
        all_points_3d: array of shape [N, 3] containing 3D coordinates
        all_colors: array of shape [N, 3] containing RGB colors (0-255)
        output_path: directory where PLY file will be saved
        filename: name of the output file (without extension)
    
    Returns:
        full_path: complete path to the created PLY file
    """
    print("\n" + "="*60)
    print("STEP 5: Exporting Point Cloud to PLY")
    print("="*60)
    
    # Ensure output directory exists
    os.makedirs(output_path, exist_ok=True)
    
    # Full path for the PLY file
    full_path = os.path.join(output_path, filename)
    
    # Use the provided ply_creator function
    print(f"\nExporting {all_points_3d.shape[0]:,} points to {filename}.ply...")
    ply_file = ply_creator(all_points_3d, all_colors, full_path)
    
    print(f" Point cloud exported successfully!")
    print(f"  File location: {ply_file}")
    print(f"  Number of points: {all_points_3d.shape[0]:,}")
    print(f"\nYou can now open '{ply_file}' in MeshLab or another 3D viewer.")
    print("="*60)
    
    return ply_file


def create_visualization_summary(all_points_3d, all_colors, camera_centers):
    """
    Create a summary visualization showing point cloud and cameras.
    """
    from mpl_toolkits.mplot3d import Axes3D
    
    fig = plt.figure(figsize=(15, 5))
    
    # Subplot 1: Point cloud sample
    ax1 = fig.add_subplot(131, projection='3d')
    
    n_points_viz = min(50000, all_points_3d.shape[0])
    step = max(1, all_points_3d.shape[0] // n_points_viz)
    points_sample = all_points_3d[::step]
    colors_sample = all_colors[::step] / 255.0
    
    ax1.scatter(points_sample[:, 0], points_sample[:, 1], points_sample[:, 2],
               c=colors_sample, marker='.', s=1, alpha=0.5)
    
    # Plot cameras
    ax1.scatter(camera_centers[:, 0], camera_centers[:, 1], camera_centers[:, 2],
               c='red', marker='o', s=100, label='Cameras')
    
    ax1.set_xlabel('X (m)')
    ax1.set_ylabel('Y (m)')
    ax1.set_zlabel('Z (m)')
    ax1.set_title(f'Point Cloud Sample\n({n_points_viz:,} of {all_points_3d.shape[0]:,} points)')
    ax1.legend()
    
    # Subplot 2: Point density histogram
    ax2 = fig.add_subplot(132)
    center = np.mean(camera_centers, axis=0)
    distances = np.linalg.norm(all_points_3d - center, axis=1)
    ax2.hist(distances, bins=50, color='steelblue', edgecolor='black', alpha=0.7)
    ax2.set_xlabel('Distance from Camera Center (m)')
    ax2.set_ylabel('Number of Points')
    ax2.set_title('Point Distribution')
    ax2.grid(True, alpha=0.3)
    
    # Subplot 3: Top-down view
    ax3 = fig.add_subplot(133)
    ax3.scatter(points_sample[:, 0], points_sample[:, 2], 
               c=colors_sample, marker='.', s=1, alpha=0.3)
    
    ax3.scatter(camera_centers[:, 0], camera_centers[:, 2], 
               c='red', marker='o', s=100, edgecolors='black', 
               linewidths=2, zorder=5, label='Cameras')
    ax3.plot(camera_centers[:, 0], camera_centers[:, 2], 
            'r--', alpha=0.5, linewidth=1)
    
    ax3.set_xlabel('X (m)')
    ax3.set_ylabel('Z (m)')
    ax3.set_title('Top-Down View')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    ax3.axis('equal')
    
    plt.tight_layout()
    plt.show()


def print_reconstruction_statistics(all_points_3d, all_colors, depth_scales, camera_centers):
    """
    Print comprehensive statistics.
    """
    print("\n" + "="*60)
    print("RECONSTRUCTION STATISTICS")
    print("="*60)
    
    print(f"\nPoint Cloud:")
    print(f"  Total points: {all_points_3d.shape[0]:,}")
    print(f"  Memory size: {all_points_3d.nbytes / (1024**2):.2f} MB (coordinates)")
    print(f"               {all_colors.nbytes / (1024**2):.2f} MB (colors)")
    
    print(f"\nSpatial Extent:")
    print(f"  X range: [{all_points_3d[:, 0].min():.3f}, {all_points_3d[:, 0].max():.3f}] m "
          f"(span: {all_points_3d[:, 0].max() - all_points_3d[:, 0].min():.3f} m)")
    print(f"  Y range: [{all_points_3d[:, 1].min():.3f}, {all_points_3d[:, 1].max():.3f}] m "
          f"(span: {all_points_3d[:, 1].max() - all_points_3d[:, 1].min():.3f} m)")
    print(f"  Z range: [{all_points_3d[:, 2].min():.3f}, {all_points_3d[:, 2].max():.3f}] m "
          f"(span: {all_points_3d[:, 2].max() - all_points_3d[:, 2].min():.3f} m)")
    
    print(f"\nCamera Setup:")
    cam_span = camera_centers.max(axis=0) - camera_centers.min(axis=0)
    print(f"  Camera trajectory span: [{cam_span[0]:.3f}, {cam_span[1]:.3f}, {cam_span[2]:.3f}] m")
    
    print(f"\nDepth Scales:")
    for i, scale in enumerate(depth_scales):
        print(f"  Camera {i}: {scale:.6f}")
    
    print("="*60)


# Run Step 5
print("\n" + "="*60)
print("STEP 5: Export Dense Point Cloud")
print("="*60)

# Print detailed statistics
print_reconstruction_statistics(all_points_3d, all_colors, depth_scales, camera_centers)

# Create visualization summary
print("\nCreating visualization summary...")
create_visualization_summary(all_points_3d, all_colors, camera_centers)

# Export to PLY file
ply_file_path = export_point_cloud_to_ply(
    all_points_3d, 
    all_colors, 
    ply_path, 
    filename='dense_reconstruction'
)

print("\n" + "="*60)
print(" PROJECT COMPLETE! ")
print("="*60)
print("\nAll steps completed successfully:")
print("   Step 1: Image undistortion")
print("   Step 2: Camera pose estimation")
print("   Step 3: Feature matching and depth scaling")
print("   Step 4: Dense 3D point reconstruction")
print("   Step 5: Point cloud export")
print(f"\nYour dense 3D reconstruction is ready!")
print(f"Open '{ply_file_path}' in MeshLab to view the result.")
print("\nRecommended MeshLab settings for best visualization:")
print("  - Render > Shading: None (to see point colors)")
print("  - Render > Color: Per Vertex")
print("  - View > Show Axis")
print("="*60)
```
